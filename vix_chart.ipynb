Quick script to show VIX for 2022 and 2023 overlayed

Data from https://www.macrotrends.net/datasets/2603/vix-volatility-index-historical-chart

```python
VIX_FILE_PATH = "data/vix_data_from_macrotrends.csv"
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# print(os.getcwd())
df = pd.read_csv(VIX_FILE_PATH)
df["Date"] = pd.to_datetime(df["Date"])

df_2022 = df[df["Date"].dt.year == 2022].copy()
df_2023 = df[df["Date"].dt.year == 2023].copy()


pass
plt.figure(figsize=(10, 6))

plt.plot(df_2022["Date"], df_2022["Value"], label="2022")
plt.plot(df_2023["Date"], df_2023["Value"], label="2023")

plt.xlabel("Date")
plt.ylabel("VIX")
plt.title("S&P 500 VIX for 2022 and 2023")
plt.legend()

plt.show()
```

```python
# Alternatively make them overlayed to compare against each other

df_2022["Day_of_year"] = df_2022["Date"].dt.day_of_year
df_2023["Day_of_year"] = df_2023["Date"].dt.day_of_year

plt.figure(figsize=(10, 6))

plt.plot(df_2022["Day_of_year"], df_2022["Value"], label="2022")
plt.plot(df_2023["Day_of_year"], df_2023["Value"], label="2023")

plt.xlabel("Day of Year")
plt.ylabel("VIX")
plt.title("S&P 500 VIX for 2022 and 2023")
plt.legend()

plt.show()
```
